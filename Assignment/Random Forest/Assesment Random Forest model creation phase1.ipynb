```python
import pandas as pd
```

```python
dataset=pd.read_csv("Insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_female', 'sex_male',
       'smoker_no', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor=RandomForestRegressor(n_estimators=100,criterion='absolute_error',max_features='log2')
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
```

```python
filename="Finalized_RandomForest_Model.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
