```python
import pandas as pd
```

```python
dataset =pd.read_csv("Insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independence=dataset[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
```

```python
dependence=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(independence,dependence,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
regressor=AdaBoostRegressor(loss='linear')
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```
